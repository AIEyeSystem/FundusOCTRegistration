```python
import os
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt

def normalize(x):
    x = 1.0*x
    return (x-np.min(x))/(np.max(x)-np.min(x))

def showMeanImg(img_stack_array):
    sum_stack = np.mean(img_stack_array,axis=1)
    # print(sum_stack.shape,np.max(sum_stack))
    sum_stack = (normalize(sum_stack)*255).astype(np.uint8)
    return sum_stack

img_stack_num = str(6).zfill(4)
img_stack_root = os.path.join('../images/images/','t'+img_stack_num)

img_stack_list = glob.glob(os.path.join(img_stack_root,'*.jpg'))
img_stack_list = sorted(img_stack_list,key=lambda x: int(os.path.split(x)[-1].split('.')[0]), reverse=False)
print(img_stack_list)
img_stack = []
for stack_id,img_name in enumerate(img_stack_list):
    img = cv2.imread(img_name)
    img_stack.append(img)

img_stack_array = np.array(img_stack)
print('img_stack_shape',img_stack_array.shape)
# print(img_stack_array.shape,np.max(img_stack_array))
### (400, 640, 400, 3) #(num,h,w,c)
# cv2.imshow('img',img_stack_array[:,200,:,:])

# norm_img_stack_array = normalize(img_stack_array)

# print(sum_stack.shape)

# img = sum_stack #img_stack_array[100]

# img_show = img_stack_array[:,200,:,:]

oct_project = showMeanImg(img_stack_array)
oct_project = cv2.resize(oct_project,(512,512))
plt.imshow(oct_project)
cv2.imwrite(img_stack_num+'.png',oct_project)
# cv2.imshow('img',img_show)
# cv2.waitKey(0)
# cv2.destroyAllWindows()


```
